```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('results_mpe.csv', delimiter=',')

for idx, heuristic in enumerate(range(3)):
    heuristic = idx + 1
    new_df = df.query('Heuristic==' + str(heuristic))[['Graph_size', 'Computation time']]
    plt.plot(new_df['Computation time'], new_df['Graph_size'], label=str(heuristic))

#add legend
plt.legend(title='Heuristic')

#add axes labels and a title
plt.ylabel('Graph size', fontsize=14)
plt.xlabel('Computation time in sec', fontsize=14)
plt.title('MPE performance evaluation', fontsize=16)

plt.show()
```
